```python
# Agrupar por plataforma y año
ventas_por_plataforma_anual = (
    games.groupby(['platform', 'year_of_release'])['global_sales']
    .sum()
    .reset_index()
    .dropna(subset=['year_of_release'])
)
# Convertir a entero
ventas_por_plataforma_anual['year_of_release'] = ventas_por_plataforma_anual['year_of_release'].astype(int)
# Filtrar 
games_filtrado = ventas_por_plataforma_anual[
    (ventas_por_plataforma_anual['year_of_release'] >= 1980) &
    (ventas_por_plataforma_anual['year_of_release'] <= 2020)
]
games_filtrado
```

```python
# Calcular el primer y último año por plataforma
vida_util = (
    games_filtrado.groupby('platform')['year_of_release']
    .agg(['min', 'max'])
    .reset_index()
    .rename(columns={'min': 'año_aparicion', 'max': 'año_ultima_venta'})
)

# Calcular duración activa
vida_util['duracion_activa'] = vida_util['año_ultima_venta'] - vida_util['año_aparicion']

# Limpiar datos
vida_util = vida_util.replace([np.inf, -np.inf], np.nan).dropna()
```

```python
# Mediana del año de aparición y promedio de duración
promedio_aparicion = vida_util['año_aparicion'].median()
promedio_duracion = vida_util['duracion_activa'].mean()

print(f"Año promedio de aparición: {promedio_aparicion}")
print(f"Duración promedio: {promedio_duracion:.1f} años")
```

```python
# Consolas con duración mayor o menor al promedio
consolas_largas = vida_util[vida_util['duracion_activa'] > promedio_duracion]
consolas_largas
```

```python
consolas_cortas = vida_util[vida_util['duracion_activa'] <= promedio_duracion]
consolas_cortas
```

```python
plt.figure(figsize=(14, 6))
plt.bar(consolas_largas['platform'], consolas_largas['duracion_activa'])
plt.title('Plataformas con duración mayor al promedio (7.6 años)')
plt.xlabel('Plataforma')
plt.ylabel('Años activos')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Paso 5.- Prueba de Hipotesis


Hipótesis nula:Las calificaciones promedio de los usuarios para los juegos en Xbox One y PC 
son iguales.(matematicamente)
Hipótesis alternativa:
Las calificaciones promedio de los usuarios para los juegos en Xbox One y PC son diferentes.
(matematicamente)


```python
# Convertir user_score a numérico
games['user_score'] = pd.to_numeric(games['user_score'], errors='coerce')

# Filtrar datos
xone_scores = games[games['platform'] == 'xone']['user_score'].dropna()
pc_scores   = games[games['platform'] == 'pc']['user_score'].dropna()

# Prueba t
alpha= 0.05
t_stat, p_value = stats.ttest_ind(xone_scores, pc_scores, equal_var=False)

# Resultados
print("Estadístico t:", t_stat)
print(f"Valor p: {p_value:.8f}")
print("Promedio XOne:", xone_scores.mean())
print("Promedio PC:", pc_scores.mean())
```

nota: se escogio el alpha 0.05 por que es el númeor mas comun.
El valor p es mucho menor, lo que indica que sí hay una diferencia significativa entre las calificaciones promedio de usuarios en Xbox One y PC.
No se puede afirmar que las calificaciones promedio son iguales.



Hipótesis nula:Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son iguales.
Hipótesis alternativa:
Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son diferentes.


```python
# Asegurarse que user_score es numérico
games['user_score'] = pd.to_numeric(games['user_score'], errors='coerce')

# Filtrar géneros 'accion' y 'deportes'
accion_scores = games[games['genre'] == 'action']['user_score'].dropna()
deportes_scores = games[games['genre'] == 'sports']['user_score'].dropna()

# Prueba de Levene para ver si las varianzas son iguales
levene_stat, levene_p = stats.levene(accion_scores, deportes_scores)
equal_var = levene_p > 0.05  # True si las varianzas son iguales

# Prueba t de dos muestras independientes
alpha = 0.05 
t_stat, p_value = stats.ttest_ind(accion_scores, deportes_scores)

# Mostrar resultados
print("=== PRUEBA DE LEVENE ===")
print("Estadístico:", levene_stat)
print("Valor p:", levene_p)
print("¿Varianzas iguales?:", equal_var)

print("\n=== PRUEBA T ===")
print("Estadístico t:", t_stat)
print("Valor p:", p_value)
print("Promedio Acción:", accion_scores.mean())
print("Promedio Deportes:", deportes_scores.mean())
```
